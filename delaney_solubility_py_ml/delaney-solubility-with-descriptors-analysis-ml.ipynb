# **Delaney solubility with descriptors analysis**

##This notebook aims to conduct a data analysis, followed by the creation of a solubility model based on Delaney's criteria using the dataset's provided attributes.

##My approach will encompass several key phases:

##- Conducting an exploratory data analysis, which is crucial for evaluating the data's integrity and particularly for pinpointing features with strong correlations.
##- Developing a stacked model.
##- Enhancing this model by fine-tuning the quantity of foundational models and the range of features involved.

#Load data

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/xluki3/data/main/delaney_solubility_with_descriptors.csv')
df
```

#Data preparation

##Data separation as X and Y

```python
y = df['logS']
y
```

```python
X = df.drop('logS', axis=1)
X
```

#Data spliting

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state= 100)
```

```python
X_train
```

```python
X_test
```

##Linear regression

###Training model

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

###Aplying the model to make a prediction

```python
y_lr_train_pred = lr.predict(X_train)
y_lr_test_pred = lr.predict(X_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

###**Evaluate model Performance**

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
print('LR MSE (Train): ', lr_train_mse)
print('LR R2 (Train): ', lr_train_r2)
print('LR MSE (Test): ', lr_test_mse)
print('LR R2 (Test): ', lr_test_r2)
```

```python
lr_results = pd.DataFrame(['Linear regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ['Method', 'Traing MSE', 'Traing R2', 'Test MSE', 'Test R2']
lr_results
```

##RandomForest

###Training the model

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth = 2, random_state = 100)
rf.fit(X_train, y_train)
```

###Applying the model to make a prediction


```python
y_rf_train_pred = rf.predict(X_train)
y_rf_test_pred = rf.predict(X_test)
```

###Evaluate model performance

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
rf_results = pd.DataFrame(['Random forest regression', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['Method', 'Traing MSE', 'Traing R2', 'Test MSE', 'Test R2']
rf_results
```

```python
df_models = pd.concat([lr_results, rf_results], axis = 0)
df_models.reset_index(drop = True)
```
